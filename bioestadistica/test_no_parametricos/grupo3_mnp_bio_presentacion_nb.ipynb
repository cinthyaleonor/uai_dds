## Tests de Muestras Independientes {.medium}

<br>

:::{.small-table}

| **Test**                      | **Descripci贸n**                                 | **Uso**                                 | **Tipo de Variables que se Comparan**               |
|-------------------------------|-------------------------------------------------|-----------------------------------------|----------------------------------------------------|
| **Chi-cuadrado**              | Eval煤a la asociaci贸n entre dos variables categ贸ricas. | An谩lisis de tablas de contingencia.    | Categ贸rica vs. Categ贸rica                           |
| **Mann-Whitney U**            | Compara dos grupos independientes en una variable continua u ordinal. | Alternativa a la prueba t para dos muestras independientes. | Continua u ordinal vs. Categ贸rica (grupos independientes) |
| **Kruskal-Wallis**            | Compara las medianas de tres o m谩s grupos independientes. | An谩lisis de preferencias entre grupos independientes. | Continua u ordinal vs. Categ贸rica (tres o m谩s grupos) |
| **Spearman**                  | Mide la correlaci贸n entre dos variables sin asumir normalidad. | An谩lisis de correlaci贸n no param茅trica. | Continua u ordinal vs. Continua u ordinal          |

:::

<br>

::: {.callout-note }

### Muestra Independiente: 
Datos de un grupo **NO est谩n relacionados** con los datos de otro grupo. Cada grupo es diferente y las observaciones en un grupo no afectan a las del otro.
:::

## Tests de Muestras Pareadas {.medium}

<br>

| **Test**                           | **Descripci贸n**                                                    | **Uso**                                       | **Tipo de Variables que se Comparan**      |
|---------------------------------------|-------------------------------------------------------------------|-----------------------------------------------|---------------------------------------------|

| **Friedman**                                     | Compara tres o m谩s grupos relacionados.                           | Alternativa a ANOVA de medidas repetidas.    | Continua u ordinal vs. Categ贸rica (grupos pareados) |

:::

<br>

::: {.callout-note}

### Muestras Pareadas 
Los datos se recogen en pares, donde cada observaci贸n de un grupo est谩 emparejada con una observaci贸n correspondiente en el otro grupo. 

Esto suele ocurrir en estudios donde se mide lo mismo en el mismo sujeto antes y despu茅s de un tratamiento, o donde se emparejan sujetos similares en diferentes condiciones. Por ejemplo: 

- Mediciones antes y despu茅s

## Caso: Estudio de Picaflores {.medium}

- Comparamos dos especies de picaflores (A y B)

- Variables de inter茅s:

    - **Frecuencia de Alimentaci贸n**: N煤mero de veces que cada especie se alimenta en un periodo determinado (*visita a flores*).

    - **Tipo de Flor Visitada**: Clasificaci贸n de las flores que eligen para alimentarse, categorizadas como Tipo A, Tipo B y Tipo C.


<br>

::: {.fragment}  

<div style="border: 2px solid #76b041; padding: 8px; background-color: #f9f9f9; font-size: 1.0em; text-align: center;">
  <strong>OBJETIVO: El objetivo de este estudio es comparar el comportamiento de dos especies de picaflores, Especie A y Especie B, en relaci贸n a su frecuencia de alimentaci贸n y el tipo de flor que visitan.</strong>
</div>

:::


````python
#| echo: false
#| output: false

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
from scipy.stats import mannwhitneyu
from scipy.stats import ttest_ind
from tabulate import tabulate

import seaborn as sns
import matplotlib.pyplot as plt

# Definir una paleta personalizada inspirada en picaflores
hummingbird_palette = ["#76b041",  # Verde esmeralda
                       "#4f83cc",  # Azul iridiscente
                       "#ff6361",  # Rojo rub铆
                       "#ffa600",  # Amarillo dorado
                       "#9b59b6"]  # Morado vibrante

# Usar la paleta personalizada en seaborn
sns.set_palette(sns.color_palette(hummingbird_palette))

def generate_hummingbird_data(n_samples, scale_a, scale_b):
    species_a = np.random.exponential(scale=scale_a, size=n_samples)
    species_b = np.random.exponential(scale=scale_b, size=n_samples)
    
    feeding_freq = np.concatenate([species_a, species_b]) + np.random.normal(0.5, 0.2, 2*n_samples)
    feeding_freq = np.round(feeding_freq).astype(int)
    feeding_freq = np.clip(feeding_freq, 1, None)
    
    flower_types = np.random.choice(['Tipo A', 'Tipo B', 'Tipo C'], 2*n_samples)
    
    df = pd.DataFrame({
        'Especie': ['A'] * n_samples + ['B'] * n_samples,
        'Frecuencia_Alimentacion': feeding_freq,
        'Tipo_Flor': flower_types
    })
    
    return df

np.random.seed(42)
df_picaflores = generate_hummingbird_data(100, 1, 3)
````

````python
#| echo: false
#| output: false

# Generar el resumen
summary = df_picaflores.groupby(['Especie', 'Tipo_Flor']).agg(
    N=('Frecuencia_Alimentacion', 'size'),
    F_A_Media=('Frecuencia_Alimentacion', 'mean'),
    Desviacion_Estandar=('Frecuencia_Alimentacion', 'std')
).reset_index()

# Mostrar el resumen en el formato solicitado
print("| Especie | Tipo de Flor | N  | *F.A.* Media    | Desviaci贸n Est谩ndar |")
print("|---------|--------------|----|------------------|----------------------|")
for index, row in summary.iterrows():
    print(f"| {row['Especie']}       | {row['Tipo_Flor']}      | {row['N']}  | {row['F_A_Media']:.1f}              | {row['Desviacion_Estandar']:.1f}                  |")
````


## Mann-Whitney U - Grupos {.smaller}

*Test no param茅trico que se utiliza para comparar dos grupos independientes cuando los datos no siguen una distribuci贸n normal.*

-   Compara dos grupos independientes
-   No requiere distribuci贸n normal
-   Eval煤a diferencias en medianas

:::{.callout-tip}
### Uso en el estudio
Comparar frecuencias de alimentaci贸n entre especies A y B
:::
::: {.callout-note}
### Hip贸tesis Nula ($H_0$)
No hay diferencia en la distribuci贸n de frecuencias de alimentaci贸n entre las dos especies



````python
#| echo: false
#| output: true

from scipy.stats import mannwhitneyu
from scipy import stats

# Filtrar los datos por especie
freq_a = df_picaflores[df_picaflores['Especie'] == 'A']['Frecuencia_Alimentacion']
freq_b = df_picaflores[df_picaflores['Especie'] == 'B']['Frecuencia_Alimentacion']

# Realizar el test de Mann-Whitney U
stat, p_value = stats.mannwhitneyu(freq_a, freq_b)
print(f'Mann-Whitney U Test: U={stat}, p={p_value:.4f}')

# Mostrar resultados
print(f'Estad铆stico U: {stat}')
print(f'Valor p: {p_value}')

# Graficar boxplot
plt.figure(figsize=(7,5))
sns.boxplot(x='Especie', y='Frecuencia_Alimentacion', data=df_picaflores)
plt.title('Comparaci贸n de Grupos A y B')
plt.ylabel('Valores')
plt.xlabel('Grupo')
plt.grid(True)

# Agregar anotaci贸n de significancia
alpha = 0.05
if p_value < alpha:
    plt.text(0.5, max(df_picaflores['Frecuencia_Alimentacion']) * 0.9, 'p < 0.05 (significativo)', ha='center', color='tomato')
else:
    plt.text(0.5, max(df_picaflores['Frecuencia_Alimentacion']) * 0.9, 'p >= 0.05 (no significativo)', ha='center', color='blue')

plt.show()
````

## Wilcoxon Signed-Rank - Antes y Despu茅s {.smaller}


*Test no param茅trico utilizado para comparar dos muestras relacionadas o medidas repetidas en un mismo grupo*

- Las diferencias entre pares son continuas y sim茅tricas

- Los datos son pareados

- No requiere normalidad en los datos

- Escala ordinal o superior

:::{.callout-tip}
### Uso en el estudio
Comparar cambios en preferencias de flores en cada especie
:::
::: {.callout-note}
### Hip贸tesis Nula ($H_0$)
La mediana de las diferencias entre pares es cero
:::


````python
#| echo: false
#| output: true

# Generar datos de ejemplo
np.random.seed(42)
n_samples = 100

# Generar datos para especie A (antes)
#df_a = generate_hummingbird_data(n_samples, scale_a=1, scale_b=3)
#df_a = df_picaflores
freq_a_antes = df_picaflores[df_picaflores['Especie'] == 'A']['Frecuencia_Alimentacion'].values

# Generar datos para especie A (despu茅s)
freq_a_despues = freq_a_antes + np.random.normal(loc=1, scale=0.5, size=n_samples)  # Aumento en frecuencia

# Generar datos para especie B (antes)
#df_b = generate_hummingbird_data(n_samples, scale_a=1.5, scale_b=2)
freq_b_antes = df_picaflores[df_picaflores['Especie'] == 'B']['Frecuencia_Alimentacion'].values

# Generar datos para especie B (despu茅s)
freq_b_despues = freq_b_antes + np.random.normal(loc=1, scale=0.5, size=n_samples)  # Aumento en frecuencia

# Crear un DataFrame para facilitar la comparaci贸n
df = pd.DataFrame({
    'Antes_A': freq_a_antes,
    'Despu茅s_A': freq_a_despues,
    'Antes_B': freq_b_antes,
    'Despu茅s_B': freq_b_despues
})

# Realizar el test de Wilcoxon para especie A
stat_a, p_value_a = stats.wilcoxon(freq_a_antes, freq_a_despues)

# Realizar el test de Wilcoxon para especie B
stat_b, p_value_b = stats.wilcoxon(freq_b_antes, freq_b_despues)

# Mostrar resultados
print(f'Especie A - Wilcoxon Test: Estad铆stico W={stat_a}, p={p_value_a:.4f}')
print(f'Especie B - Wilcoxon Test: Estad铆stico W={stat_b}, p={p_value_b:.4f}')

# Graficar boxplot para ambas especies
plt.figure(figsize=(7, 5))

# Crear una lista de datos para el boxplot
data_to_plot = [freq_a_antes, freq_a_despues, freq_b_antes, freq_b_despues]
labels = ['Antes A', 'Despu茅s A', 'Antes B', 'Despu茅s B']

sns.boxplot(data=data_to_plot)
plt.title('Comparaci贸n Antes y Despu茅s del Tratamiento')
plt.ylabel('Frecuencia de Alimentaci贸n')
plt.xlabel('Condici贸n')
plt.xticks(ticks=range(len(labels)), labels=labels)
plt.grid(True)

# Agregar anotaci贸n de significancia para especie A
alpha = 0.05
if p_value_a < alpha:
    plt.text(0.5, max(freq_a_despues) * 0.95, 'p < 0.05 \n(significativo)', ha='center', color='tomato')
else:
    plt.text(0.5, max(freq_a_despues) * 0.95, 'p >= 0.05\n (no significativo)', ha='center', color='blue')

# Agregar anotaci贸n de significancia para especie B
if p_value_b < alpha:
    plt.text(3, max(freq_b_despues) * 0.95, 'p < 0.05\n (significativo)', ha='center', color='tomato')
else:
    plt.text(3, max(freq_b_despues) * 0.95, 'p >= 0.05\n (no significativo)', ha='center', color='blue')

plt.show()
````

## Friedman Test - Preferencias {.smaller}

*Busca comparar 3 o m谩s grupos relacionados cuando los datos no cumplen con los supuestos de normalidad.*

- Datos ordinales o continuos
- Muestras relacionadas o medidas repetidas
- No requiere normalidad
- Bloques completos aleatorios

:::{.callout-tip}
### Uso en el estudio
Medir la frecuencia de alimentaci贸n de un grupo de picaflores en diferentes tipos de flores (*las flores son vistitadas por picaflores que se repiten en las visitas*).
:::
::: {.callout-note}
### Hip贸tesis Nula ($H_0$)
La mediana de las frecuencias de visitas a las diferentes flores es igual para todas las especies de picaflores.
:::

````python
#| echo: false
#| output: true

# Crear una tabla din谩mica
pivot_table = df_picaflores.pivot_table(values='Frecuencia_Alimentacion', 
                                         index='Especie', 
                                         columns='Tipo_Flor', 
                                         aggfunc='mean')

# Realizar el Friedman Test
stat, p_value = stats.friedmanchisquare(*[pivot_table[col] for col in pivot_table.columns])

# Mostrar resultados
print(f'Friedman Test: Estad铆stico = {stat}, p = {p_value:.4f}')

# Graficar el heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title(f'Preferencias de Alimentaci贸n por Especie y Tipo de Flor\nFriedman Test: p={p_value:.4f}')
plt.ylabel('Especie')
plt.xlabel('Tipo de Flor')
plt.show()

# Interpretar resultados
if p_value < 0.05:
    print("Hay diferencias significativas entre las preferencias de alimentaci贸n por tipo de flor.")
else:
    print("No hay diferencias significativas entre las preferencias de alimentaci贸n por tipo de flor.")
````

## Kruskal-Wallis Test -  Grupos Independientes {.smaller}

*Busca determinar si hay diferencias estad铆sticamente significativas entre 3 o m谩s grupos independientes.*

- Comparar las medianas de 3 o m谩s grupos independientes
- Extensi贸n de Mann-Whitney U
- Es considerado la versi贸n no param茅trica del ANOVA de una v铆a

:::{.callout-tip}
### Uso en el estudio
Determinar si hay diferencias significativas en la frecuencia de alimentaci贸n entre los tipos de flores que visitan los picaflores
:::
::: {.callout-note}
### Hip贸tesis Nula ($H_0$)
 Las medianas de la frecuencia de alimentaci贸n entre los diferentes tipos de flores (Tipo A, Tipo B y Tipo C) son iguales para las especies de picaflores A y B.
:::


````python
# Realizar el test de Kruskal-Wallis
species = df_picaflores['Especie'].unique()
flower_visits = [df_picaflores[df_picaflores['Especie'] == sp]['Frecuencia_Alimentacion'] for sp in species]

h_statistic, p_value = stats.kruskal(*flower_visits)
# Preparar datos para la visualizaci贸n
flower_counts = df_picaflores.groupby(['Especie', 'Tipo_Flor'])['Frecuencia_Alimentacion'].sum().unstack(fill_value=0)

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(7, 6))

# Graficar en los ejes especificados
flower_counts.plot(kind='barh', width=0.8, ax=ax)

# Personalizar el gr谩fico
ax.set_title(f'Preferencia de Flores por Especie\nKruskal-Wallis: H={h_statistic:.2f}, p={p_value:.4f}')
ax.set_ylabel('Especie')
ax.set_xlabel('N煤mero de Visitas')
ax.grid(axis='x', linestyle='--', alpha=0.7)  # Agregar grid para mejor legibilidad

# Ajustar leyenda
ax.legend(title='Tipo de Flor', bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=3)  # Leyenda centrada debajo del gr谩fico

# Ajustar m谩rgenes
plt.tight_layout()

# Mostrar el gr谩fico
plt.show()
# Imprimir resultados
#print(f"Estad铆stico H de Kruskal-Wallis: {h_statistic:.4f}")
#print(f"Valor p: {p_value:.4f}")

if p_value < 0.05:
    print("Hay diferencias significativas entre al menos dos especies.")
else:
    print("No hay evidencia suficiente para concluir que existen diferencias.")
````

## An谩lisis Comparativo de M茅todos Estad铆sticos {.smaller}

**И Pruebas Realizadas:**

1. **Comparaci贸n Entre Especies:**
   - T-test (param茅trico)
   - Mann-Whitney U (no param茅trico)

2. **Comparaci贸n Entre Tipos de Flores:**
   - Friedman (no param茅trico)
   - Kruskal-Wallis (no param茅trico)

** Resultados:**

- T-test: Infiere significancia a pesar de la falta de normalidad, lo que puede ser enga帽oso.
- Mann-Whitney U, Friedman, y Kruskal-Wallis: Son m谩s apropiados y confiables en este contexto.

````python
#| echo: false
#| output: false

import pandas as pd
from scipy import stats
import numpy as np
from statsmodels.stats.power import TTestIndPower, FTestAnovaPower
from tabulate import tabulate

# Simulaciones
np.random.seed(42)
n_samples = 30

# Asegurarnos de que hay datos repetidos por especie y tipo de flor
df_picaflores_extra = pd.concat([df_picaflores] * 3, ignore_index=True)

# Filtrar el DataFrame para asegurarnos de que tenemos un n煤mero igual de observaciones
df_grouped = df_picaflores_extra.groupby('Tipo_Flor')

# Asegurarse de que cada tipo de flor tenga la misma cantidad de datos
min_size = df_grouped.size().min()  # Encuentra el tama帽o m铆nimo de los grupos
data = [group['Frecuencia_Alimentacion'].sample(min_size).values for name, group in df_grouped]

# Test de Friedman (no param茅trico)
f_stat, f_p_value = stats.friedmanchisquare(*data)

# T-test (param茅trico) y Mann-Whitney U test (no param茅trico)
species_a = df_picaflores_extra[df_picaflores_extra['Especie'] == 'A']['Frecuencia_Alimentacion'].sample(min_size).values
species_b = df_picaflores_extra[df_picaflores_extra['Especie'] == 'B']['Frecuencia_Alimentacion'].sample(min_size).values
t_stat, t_p_value = stats.ttest_ind(species_a, species_b)
u_stat, u_p_value = stats.mannwhitneyu(species_a, species_b)

# Kruskal-Wallis H-test (no param茅trico)
h_stat, h_p_value = stats.kruskal(*data)

# Calcular el poder
power_t = TTestIndPower().power(effect_size=0.5, nobs1=n_samples, alpha=0.05)  # Ejemplo de efecto de tama帽o 0.5

# Crear tabla de resultados
results = [
    ["T-test", "Param茅trico", f"{t_stat:.4f}", f"{t_p_value:.4f}", f"{power_t:.4f}"],
    ["Mann-Whitney U", "No Param茅trico", f"{u_stat:.4f}", f"{u_p_value:.4f}", "N/A"],
    ["Friedman", "No Param茅trico", f"{f_stat:.4f}", f"{f_p_value:.4f}", "N/A"],
    ["Kruskal-Wallis H", "No Param茅trico", f"{h_stat:.4f}", f"{h_p_value:.4f}", "N/A"]
]

# Mostrar tabla
#print(tabulate(results, headers=["Test", "Tipo", "Estad铆stico", "p-valor", "Poder"], tablefmt="pipe"))
````

## Pseudoreplicaci贸n y M茅todos No Param茅tricos {.smaller}

- Las mediciones de datos no son independientes, pero se tratan como si lo fueran.

- **Ejemplo**: Medir repetidamente la frecuencia de alimentaci贸n del mismo picaflor en diferentes momentos del d铆a.



````python
#| echo: false
#| output: true

np.random.seed(42)
n_birds = 5
n_measurements = 10

bird_effects = np.random.normal(0, 2, n_birds)
time_effects = np.linspace(-1, 1, n_measurements)

data = []
for bird in range(n_birds):
    for time in range(n_measurements):
        feeding_freq = 10 + bird_effects[bird] + time_effects[time] + np.random.normal(0, 0.5)
        data.append({'Picaflor': f'Picaflor {bird+1}', 'Tiempo': time, 'Frecuencia': feeding_freq})

df_pseudo = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
for bird in df_pseudo['Picaflor'].unique():
    bird_data = df_pseudo[df_pseudo['Picaflor'] == bird]
    plt.plot(bird_data['Tiempo'], bird_data['Frecuencia'], 'o-', label=bird)

plt.xlabel('Tiempo')
plt.ylabel('Frecuencia de Alimentaci贸n')
plt.title('Pseudoreplicaci贸n: Mediciones Repetidas por Picaflor')
plt.legend()
plt.show()
````

**Problema**: Los m茅todos no param茅tricos est谩ndar asumen independencia entre observaciones.

    - Inflaci贸n del tama帽o de muestra aparente

    - Aumento del riesgo de error Tipo I (falsos positivos)

**Soluciones**:

    - Usar m茅todos que consideren la dependencia (ej. modelos mixtos)

    - Agregar datos por unidad experimental antes del an谩lisis
    
    - Usar m茅todos de permutaci贸n que respeten la estructura de los datos

## Ejemplo: Manejo de Pseudoreplicaci贸n {.smaller}

Comparemos un enfoque *incorrecto* versus un enfoque *correcto*:


````python
#| echo: false
#| output: true

# Enfoque incorrecto (ignorando pseudoreplicaci贸n)
incorrect_stat, incorrect_p = stats.mannwhitneyu(
    df_pseudo[df_pseudo['Picaflor'].isin(['Picaflor 1', 'Picaflor 2'])]['Frecuencia'],
    df_pseudo[df_pseudo['Picaflor'].isin(['Picaflor 3', 'Picaflor 4', 'Picaflor 5'])]['Frecuencia']
)

# Enfoque correcto (agregando datos por picaflor)
df_aggregated = df_pseudo.groupby('Picaflor')['Frecuencia'].mean().reset_index()
correct_stat, correct_p = stats.mannwhitneyu(
    df_aggregated[df_aggregated['Picaflor'].isin(['Picaflor 1', 'Picaflor 2'])]['Frecuencia'],
    df_aggregated[df_aggregated['Picaflor'].isin(['Picaflor 3', 'Picaflor 4', 'Picaflor 5'])]['Frecuencia']
)

print(f"Enfoque incorrecto p-valor: {incorrect_p:.4f}")
print(f"Enfoque correcto p-valor: {correct_p:.4f}")

plt.figure(figsize=(7, 5))
sns.boxplot(x='Picaflor', y='Frecuencia', data=df_pseudo)
plt.title('Distribuci贸n de Frecuencias por Picaflor')
plt.show()
````

**Incorrecto**: Ignorar la dependencia entre mediciones, lo que lleva a un an谩lisis potencialmente err贸neo.

**Correcto**: Agrupar los datos para reflejar la variabilidad entre los picaflores, asegurando que se cumple la independencia de las observaciones.

- El enfoque incorrecto produce un p-valor artificialmente bajo

- El enfoque correcto proporciona un resultado m谩s conservador y preciso
